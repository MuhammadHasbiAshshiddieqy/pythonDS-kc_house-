### Find Categorical and Numerical

If you find some unusual visualization above, it means that you are not separate between data **categorical** and **numerical**.

Remembered it? Yeah, while in some visualization is using categorical data only, numerical data only, or both of them.

Let's divide them into categorical and numerical. What columns is categorical and what kind is numerical?
- data_cat for data categorical
- data_num for data numerical

```python
df.head()
```

```python
# Type here for answer

data_cat = df[['waterfront', 'condition']]
```

```python
data_cat.head()
```

```python
data_num = df.drop(['waterfront', 'condition', 'price'], axis=1)
```

```python
data_num.head()
```

#### Numerical Data Normalization
It's your turn to normalize your data for better understanding of it. Now it's numerical time.

```python
# Type here for normalize

from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()

data_normal_num = min_max_scaler.fit_transform(data_num)
data_normal_num = pd.DataFrame(data_normal_num, columns=data_num.columns)
```

```python
data_normal_num.head()
```

#### Categorical Data Normalization
Now for categorical data, let's normalize it !

```python
# Type here for normalize

from sklearn.preprocessing import LabelEncoder
Lencoder = LabelEncoder()

data_cat_norm = data_cat.copy()
length = data_cat_norm.shape[1]
col = data_cat_norm.columns
for i in range (length):
    a = Lencoder.fit_transform(data_cat_norm.iloc[:,i:i+1])
    a = pd.DataFrame(a, columns=[col[i]+'new'])
    data_cat_norm = data_cat_norm.join(a)
```

```python
data_cat_norm.head(10)
```

#### Combined and Save Your Data

After you normalize categorical and numerical, don't forget to combined all of your data and save it into csv format file.

```python
# Type here
data_norm = data_normal_num.copy()
data_norm = data_norm.join(data_cat_norm[['waterfrontnew', 'conditionnew']])
data_norm = data_norm.join(df[['price']])
```

```python
data_norm.head(10)
```

### Linear Model

Define what is your $x$ and $y$ and split into training and testing split.
For :
- One variable $x$
- Many variable $x$
- All variable $x$

```python
# Type here
Xone = data_norm[['sqft_living']] #dipilih berdasarkan korelasi
Xmany = data_norm[['sqft_living', 'grade', 'sqft_above']] #dipilih berdasarkan korelasi
Xall = data_norm.drop(['price'], axis = 1)
y = data_norm['price']
```

```python
Xone.head()
```

```python
Xmany.head()
```

```python
Xall.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
Xone_train, Xone_test, yone_train, yone_test = train_test_split(Xone,y,test_size=0.25,random_state = 45)
Xmany_train, Xmany_test, ymany_train, ymany_test = train_test_split(Xmany,y,test_size=0.25,random_state = 45)
Xall_train, Xall_test, yall_train, yall_test = train_test_split(Xall,y,test_size=0.25,random_state = 45)
```

After that, define function for Linear Model, fit and predict

```python
from sklearn.linear_model import LinearRegression
```

```python
model_reg_one = LinearRegression()
model_reg_many = LinearRegression()
model_reg_all = LinearRegression()
```

```python
model_reg_one.fit(Xone_train, yone_train)
model_reg_many.fit(Xmany_train, ymany_train)
model_reg_all.fit(Xall_train, yall_train)
```

Get the formula for each model and score of them.

**Example**:
price = 10 * grade + 5 * condition

```python
print("one feature")
print("Price = ", model_reg_one.coef_[0]," * ", Xone.columns[0]," + ", model_reg_one.intercept_)
```

```python
print("many feature")
formula = ""
for i in range(len(Xmany.columns)):
    formula = formula + str(model_reg_many.coef_[i])+" * "+ Xmany.columns[i] + " + "
print("Price = ",formula, model_reg_one.intercept_)
```

```python
print("all feature")
formula = ""
for i in range(len(Xall.columns)):
    formula = formula + str(model_reg_all.coef_[i])+" * "+ Xall.columns[i] + " + "
print("Price = ",formula, model_reg_all.intercept_)
```

```python
#predict one feature

price_one_pred = model_reg_one.predict(Xone_test)
pd.DataFrame({'actual':yone_test,'prediction':price_one_pred}).head()
```

```python
#predict many feature

price_many_pred = model_reg_many.predict(Xmany_test)
pd.DataFrame({'actual':ymany_test,'prediction':price_many_pred}).head()
```

```python
#predict all feature

price_all_pred = model_reg_all.predict(Xall_test)
pd.DataFrame({'actual':yall_test,'prediction':price_all_pred}).head()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# score for one feature
print("one feature")
print("MSE : ",mean_squared_error(yone_test, price_one_pred))
print("MAE : ",mean_absolute_error(yone_test, price_one_pred))
print("R2 : ",r2_score(yone_test, price_one_pred))
```

```python
# score for many feature
print("many feature")
print("MSE : ",mean_squared_error(ymany_test, price_many_pred))
print("MAE : ",mean_absolute_error(ymany_test, price_many_pred))
print("R2 : ",r2_score(ymany_test, price_many_pred))
```

```python
# score for all feature
print("all feature")
print("MSE : ",mean_squared_error(yall_test, price_all_pred))
print("MAE : ",mean_absolute_error(yall_test, price_all_pred))
print("R2 : ",r2_score(yall_test, price_all_pred))
```
